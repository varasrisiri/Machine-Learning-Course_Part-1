#Manipulating a Dataframe

```python
# adding a column to a Dataframe
diabetes_df = pd.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
diabetes_df['Outcome'] = diabetes_dataset.target
```

```python
diabetes_df.head()
```

```python
#removing a particular row
diabetes_df.drop(index=0,axis=0) ## for removing a row we have to give axis = 0 ; and for removing a particular column we have to give axis = 1.
```

```python
#drop a column
diabetes_df.drop(columns ='age', axis =1) # for removing a column axis = 1
```

```python
#locating a row using the index value
diabetes_df.iloc[3]
```

```python
#locating a particular column
print(diabetes_df.iloc[:, 0])     #1st column
print(diabetes_df.iloc[:, 1])     #2nd column
print(diabetes_df.iloc[:, 2])    # 3rd column
print(diabetes_df.iloc[:, -1])    #last column
```
