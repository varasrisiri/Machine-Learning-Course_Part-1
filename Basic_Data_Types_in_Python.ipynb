#Conversion of one data type to another

```python
# int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
# float to int
x = 5.83
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```

#Slicing

```python
my_string = "Programming"
print(my_string[1:5])  # values from index 1 to 5-1 will be sliced
```

```python
#step
print(my_string[0:10:2]) # values from 0 to 10-1 will be sliced but 0:10:2 (2 over there means every second letter will be removed)
```
