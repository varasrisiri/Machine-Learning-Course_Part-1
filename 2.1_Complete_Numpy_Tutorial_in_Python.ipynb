#List vs Numpy -- Time Taken

```python
from time import process_time
```

#Time taken by a list

```python
python_list = [i for i in range(500)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(500)])

start_time = process_time()

np_array += 5

end_time = process_time()

print(end_time - start_time)
```

#Initial Placeholders in numpy arrays


```python
#create a numpy array of Zeros
x = np.zeros((4,5))
print(x)
```

```python
#create a numpy array of ones
y = np.ones((3,3))
print(y)
```

```python
#array of a particular value
z = np.full((5,4), 5)
print(z)
```

```python
# create an identity matrix --> have no. of rows = no. of columns
a = np.eye(5)
print(a)
```

```python
#create a numpy array with random values
b = np.random.random((3,5))
print(b)
```

```python
#create a random interger values array with in a specific range
c = np.random.randint(10,100,(3,5))
print(c)
```

```python
# array of evenly spaced values --> specifying the number of values required
d = np.linspace(10,30,6) #here 6 is mentioning no. of values required between 10 to 30
print(d)
```

```python
# array of evenly spaced values --> specifying the step
e = np.arange(10,30,5) # here 5 is mentioning the gap between the numbers
print(e)
```

```python
# convert a list to a numpy array
list2 = [10,20,30,40,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

```python
# convert a tuple to a numpy array
tuple1 = (10,20,30,40,50)

np_array = np.array(tuple1)
print(np_array)
type(np_array)
```

#Mathematical operations on a np array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1 + list2)  #concatenate or joins two list
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))

print(a)
print(b)
```

```python
print(a+b)  ## incase of numpy array we have element wise mathematical operations
print(a-b)
print(a*b)
print(a/b)
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))

print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

#Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
## Example 2
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

#Reshaping a array

```python
a = np.random.randint(0,10,(3,4))
print(a)
print(a.shape)
```

```python
b = a.reshape(4,3)
print(b)
print(b.shape)
```
