#Imbalanced Dataset

A dataset with an unequal class distribution

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#determine the distribution of two classes
credit_card_data['Class'].value_counts()
```

0 --> legit Transactions - there's almost 99% of data

1 --> fradulent transaction - there's only 1% or less than 1% of data


This is higly imbalanced dataset --> because of most of the transactions is on one side





```python
#separating the legit and fradulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under- sampling


    Build a sample dataset containing similar distribution o flegit and fradulent transaction

Number of Fraudulent Transaction --> 492

```python
legit_sample = legit.sample(n=len(fraud))
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames


```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
