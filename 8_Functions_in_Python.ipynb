Factorial of a Number

Factorial of a number is the product of all the positive integers less than or equal to the given number

factorial of 5 = 5 × 4 × 3 × 2 × 1 = 120

```python
number = int(input("Enter a number to find its factorial : ")) # number = 5

factorial = 1

if number == 0:
  print("The factorial of 0 is 1")
else:
  for i in range(1,number+1):  #range(1,6) --> 1,2,3,4,5
    factorial = factorial*i  # --> 1*2*3*4*5
  print('The factorial of', number, 'is', factorial)
```

#Factorial Function

```python
def factorial_value(num):
  factorial = 1
  if num == 0:
    return factorial
  else:
    for i in range(1,num+1):
      factorial = factorial*i
    return factorial

print(factorial_value(int(input("enter a number to find the factorial"))))



```
