Operators in Python
1. Arithmetic Operators
2. Assignment Operators
3. Comparison Operators
4. Logical Operators
5. Identity Operators
6. Membership Operators


#Arithmetic Operators

```python
num_1 = 20
num_2 = 10

#addition
sum = num_1 + num_2
print('sum = ', sum)

#subtraction
diff = num_1 -  num_2
print('diff = ', diff)

#multiplication
mul = num_1*num_2
print('mul = ', mul)

#division
div = num_1/num_2
print('div = ', div)

#exponent
exp = num_1**num_2
print("exponent = ", exp)

#modulus
mod = num_1%num_2
print('remainder = ', mod)
```

#Assignment Operators
1. +=
2. -=
3. *=
4. **=
5. /=
6. %=

```python
a = 5
print(a)
```

```python
a = 5
a += 5 #a = a+5
print(a)
```

```python
b = 5
b -= 3 # b = 5-3
print(b)
```

#Comparison Operator

```python
a = 5
b = 10

print(a == b) # equal to
print(a != b) #not equal to
print(a > b) #greater than
print(a < b) #less than
print(a <= b)
print(a >= b)
```

#Logical Operators
1. and
2. or
3. not

```python
a = 10

print(a>20 and a>5)
print(a>20 or a>5)
print(not(a>8 and a>5))
```

#Identity Operator
1. is
2. is not

```python
x = 5
y = 10

print(x is y)
```

```python
x = 5
y = 10

print(x is not y)
```

#Membership Operator
1. in
2. not in

```python
a = 5
b = 10

c = [1,2,3,4,5,6,9]
print(a in c)
print(b in c)
```

```python
a = 5
b = 10

c = [1,2,3,4,5,6,9]
print(a not in c)
print(b not in c)
```
